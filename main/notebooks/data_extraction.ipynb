# Cleaning the main json file with actual labels of the Dataset

```python
import json

# Load the JSON data from a file
with open('../assets/classes.json', 'r') as file:
    data = json.load(file)

# Process each key-value pair
new_data = {}
for key, value in data[0].items():
    # Extract the first two digits after 'n'
    new_key = key[1:3]
    
    # Replace the whole value with those two digits
    new_data[new_key] = value

# Save the updated data back to a JSON file
with open('../assets/final_classes.json', 'w') as file:
    json.dump(new_data, file, indent=4)

print("Data has been updated and saved to new_data.json")
```
